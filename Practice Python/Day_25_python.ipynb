```python
import PyPDF2 
```

```python
path = "pypdf.pdf"
with open (path,"rb") as file :
    reader = PyPDF2.PdfReader(file)

    fast_page = reader.pages[3].extract_text()
    print(fast_page)
```

```python
path = "pypdf.pdf"
with open (path,"rb") as file :
    reader = PyPDF2.PdfReader(file)

    for page in reader.pages:
        text = page.extract_text()
        print(text)
        print("\n")
```
