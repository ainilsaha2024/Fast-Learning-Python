```python
names= ["sapnil","niloy","saha","diya","bhuti","nil"]#list var 1
age =[10,20,30,40,50,60]#list var 2 
city= ["dhaka","khulna","rajshahi","borishal","sylhet","chatogram"] #list var 3 
zipped=zip(names,city,age)# zip function use for 3 list var 
print(list(zipped))#print the zipped use list function 
```

```python
zipped=zip(names,city,age)#zip function use for 3 list var
print(tuple(zipped))#print the zipped var use tuple function 
```

```python
zipped=zip(names,city,age)#zip function use for 3 list var
print(set(zipped))#print the ziiped var use set functiuon 
```

```python
zipped=dict(zip(names,city))#zip function use for 2 var bcz dict use keys and values 
print(zipped)
```

```python
zipped.keys()#print the keys 
```

```python
zipped.values()#print the vlues 
```

# Unzip = zip(*var) 

```python
name1 ,city1,age1 =zip(*zip(names,city,age)) # unzip function use new 3 var 
print(name1)#and new var print
print(city1)
print(age1) 
```

# use loop zip file dictionary item() 

```python
zipped = {'sapnil': 'dhaka', 'niloy': 'khulna', 'saha': 'rajshahi', 'diya': 'borishal', 'bhuti': 'sylhet', 'nil': 'chatogram'}
for key, value in zipped.items():
    print(f"Name: {key}, City: {value}")
```
