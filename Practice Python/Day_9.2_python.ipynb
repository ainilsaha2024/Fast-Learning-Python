```python
# copy ( ) mathods set 
s = {1,2,3,4,5,6,7}
print(s)

```

```python
nil = s.copy() # nil var copy() set mathods
```

```python
s.remove(7) #remove( ) 7 use set remove mathods
```

```python
print(s) # show result removed 7 element
```

```python
print(nil) # but copy nil var and show again 7 element bcz fast s.copy( ) use copy mathods 
```

```python
# difference( ) mathods set 

print(s) # s var 7 element remove use to remove mathods
print(nil) # nil var 7 element use to copy mathods 

nil.difference(s) #look differnce 7 element
```

```python
d1 ={ "nil ","niloy",10,20,30}
d2 = {"sapnil","saha",10,20}
D1 = frozenset(d1) #set d1 var covert to frozenset()
D2 = frozenset(d2) #set d2 var convert to frozenset()
print(type(D1)) 
print(type(D2))
print(D1)
print(D2)
```

```python
# difference ( ) mathods 
D1.difference(D2) # D1 = "nil","niloy" but D2 ="saha","sapnil"
```

```python
D2.difference(D1) #D2 ="saha","sapnil" but D1 ="nil","niloy"
```

```python
# symmetric_differnce ( )
num = { 1,2,7,8}
num2 ={1,2,4,5}
# symmentic_difference mean common element out uncommon element output
num.symmetric_difference(num2)
```

```python
# symmentric_update ( )
num2.symmetric_difference_update(num) # num2 var element 4,5 update num 7,8 
```
